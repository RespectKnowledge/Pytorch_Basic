```
import torch.functional as F
# identity_block
class I_block(nn.Module):
  def __init__(self,in_c,out_c,stride=1):
    super(I_block,self).__init__()
    # block 1
    self.con1=nn.Conv2d(in_c,out_c,kernel_size=3,stride=stride,padding=1)
    self.bn1=nn.BatchNorm2d(out_c)
    self.rel1=nn.ReLU()
    self.conv2=nn.Conv2d(out_c,out_c,kernel_size=3,stride=stride,padding=1)

  def forward(self,x):
    identitiy=x
    x=self.con1(x)
    x=self.bn1(x)
    x=self.rel1(x)
    x=self.conv2(x)
    x += identitiy
    x = self.rel1(x)
    return x
block=I_block(in_c=64,out_c=64)
print(block)
inp=torch.rand(1,64,224,224)
out=block(inp)
print(out.shape)
```

```
import torch.functional as F
# identity_block
class Residual_block(nn.Module):
  def __init__(self,in_c,out_c,stride=1):
    super(Residual_block,self).__init__()
    # layer1 1x1
    self.con1=nn.Conv2d(in_c,out_c,kernel_size=1,stride=stride,padding=0)
    self.bn1=nn.BatchNorm2d(out_c)
    self.rel1=nn.ReLU()

    self.con2=nn.Conv2d(out_c,out_c,kernel_size=3,stride=stride,padding=1)
    self.bn2=nn.BatchNorm2d(out_c)
    self.rel2=nn.ReLU()

    self.con3=nn.Conv2d(out_c,out_c*4,kernel_size=1,stride=stride,padding=0)
    self.bn3=nn.BatchNorm2d(out_c*4)
    self.rel3=nn.ReLU()

  def forward(self,x):
    identitiy=x
    #print(identitiy.shape)
    x=self.con1(x)
    #print(x.shape)
    x=self.bn1(x)
    x=self.rel1(x)

    x=self.con2(x)
    x=self.bn2(x)
    x=self.rel2(x)

    x=self.con3(x)
    x=self.bn3(x)
    #x=self.rel3(x)
    #print(x.shape)

    x += identitiy
    x = self.rel1(x)
    return x
    
block=Residual_block(in_c=256,out_c=64)
print(block)
inp=torch.rand(1,256,224,224)
out=block(inp)
print(out.shape)
```

```
import torch.functional as F
# identity_block
class Residualtrans_block(nn.Module):
  def __init__(self,in_c,out_c,stride=2):
    super(Residualtrans_block,self).__init__()

    # skip block 1x1 layer
    self.block=nn.Sequential(nn.Conv2d(in_c,out_c*4,kernel_size=1,stride=stride,padding=0),
                             nn.ReLU())
    # layers block 1x1,128,3x3,128,1x1,512
    # layer1 1x1
    self.con1=nn.Conv2d(in_c,out_c,kernel_size=1,stride=2,padding=0)
    self.bn1=nn.BatchNorm2d(out_c)
    self.rel1=nn.ReLU()

    self.con2=nn.Conv2d(out_c,out_c,kernel_size=3,stride=1,padding=1)
    self.bn2=nn.BatchNorm2d(out_c)
    self.rel2=nn.ReLU()

    self.con3=nn.Conv2d(out_c,out_c*4,kernel_size=1,stride=1,padding=0)
    self.bn3=nn.BatchNorm2d(out_c*4)
    self.rel3=nn.ReLU()

  def forward(self,x):
    blockskip=self.block(x)
    #print(blockskip.shape)
    x=self.con1(x)
    #print(x.shape)
    x=self.bn1(x)
    x=self.rel1(x)

    x=self.con2(x)
    x=self.bn2(x)
    x=self.rel2(x)
    #print(x.shape)
    x=self.con3(x)
    x=self.bn3(x)
    #x=self.rel3(x)
    #print(x.shape)

    x += blockskip
    x = self.rel1(x)
    return x
block=Residualtrans_block(in_c=256,out_c=128)
print(block)
inp=torch.rand(1,256,224,224)
out1=block(inp)
print(out1.shape)
```

```
# ResNet50 model
class ResNet50(nn.Module):
  def __init__(self,imge_dim,n_classes):
    super(ResNet50,self).__init__()
    self.c1=nn.Conv2d(imge_dim,256,kernel_size=3,stride=2,padding=3)
    self.bn1=nn.BatchNorm2d(256)
    self.rel1=nn.ReLU()
    self.pool1=nn.MaxPool2d(kernel_size=3,stride=2)
    # block1 
    self.block11=Residual_block(256,64)
    self.block12=Residual_block(256,64)
    self.block13=Residual_block(256,64)
    # block2
    self.block21=Residualtrans_block(256,128)
    self.block22=Residual_block(512,128)
    self.block23=Residual_block(512,128)
    self.block24=Residual_block(512,128)
    # block3
    self.block31=Residualtrans_block(512,256)
    self.block32=Residual_block(1024,256)
    self.block33=Residual_block(1024,256)
    self.block34=Residual_block(1024,256)
    self.block35=Residual_block(1024,256)
    self.block36=Residual_block(1024,256)
    # block4
    self.block41=Residualtrans_block(1024,512)
    self.block42=Residual_block(2048,512)
    self.block43=Residual_block(2048,512)

    self.avgpool=nn.AdaptiveAvgPool2d((1,1))
    self.fc=nn.Linear(2048,n_classes)

  def forward(self,x):
    x=self.c1(x)
    x=self.bn1(x)
    x=self.rel1(x)
    x=self.pool1(x)
    #print(x.shape)
    x=self.block11(x)
    print(x.shape)
    x=self.block12(x)
    print(x.shape)
    x=self.block13(x)
    print(x.shape)
    x=self.block21(x)
    x=self.block22(x)
    x=self.block23(x)
    x=self.block24(x)
    x=self.block31(x)
    x=self.block32(x)
    x=self.block33(x)
    x=self.block34(x)
    x=self.block35(x)
    x=self.block36(x)
    print(x.shape)
    x=self.block41(x)
    x=self.block42(x)
    x=self.block43(x)
    x=self.avgpool(x)
    x=x.reshape(x.shape[0],-1)
    x=self.fc(x)
    print(x.shape)
    return x

model=ResNet50(3,1000)
inp=torch.rand(1,3,224,224)
out=model(inp)
#print(out.shape)
#print(model)
```
